```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import control
```

```python
def complete_tank_system(t, h, params):
    g, A1, A2, A3, a2, a3, a4, a5, Smax, u = params
    h1, h2, h3 = h
    dhdt = np.zeros(3)
    dhdt[0] = (u - Smax * a2 * np.sqrt(2 * g * h1) - Smax * a3 * np.sqrt(2 * g * h1)) / A1
    dhdt[1] = (Smax * a2 * np.sqrt(2 * g * h1) - Smax * a4 * np.sqrt(2 * g * h2)) / A2
    dhdt[2] = (Smax * a3 * np.sqrt(2 * g * h1) - Smax * a5 * np.sqrt(2 * g * h3)) / A3
    
    return dhdt
```
