```python

def svd_compression_demo(img, compression_ratio):
    U, S, Vt = np.linalg.svd(img, full_matrices=False)
    total_components = len(S)
    k = max(1, int(total_components * compression_ratio))
    compressed = U[:, :k] @ np.diag(S[:k]) @ Vt[:k, :]
    original_size = img.shape[0] * img.shape[1]
    compressed_size = (U.shape[0]-k)*( Vt.shape[1]-k)
    print('compressed size ' , compressed_size)
    print('original size ' , original_size)


    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f"Compressed (k={k})")
    plt.imshow(compressed, cmap='gray')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    
svd_compression_demo(img , 0.05)

```
